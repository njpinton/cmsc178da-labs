## Part 3: Diagnostics & Assumptions

A regression is only as reliable as its assumptions. The **LINE** mnemonic covers the four key checks:
- **L**inearity — relationship between X and Y is linear
- **I**ndependence — residuals are independent of each other
- **N**ormality — residuals are approximately normally distributed
- **E**qual variance — residual spread is constant across fitted values (homoscedasticity)

## Dataset 2: Philippine Credit Scoring (Simulated)

For logistic regression, we generate a realistic credit scoring dataset inspired by GCash GLoan and BSP lending data. Each row represents a loan applicant.

```python
np.random.seed(42)
n = 1000

monthly_income = np.random.lognormal(mean=10.2, sigma=0.5, size=n)  # PHP ~25K median
outstanding_balance = np.random.exponential(scale=50000, size=n)
employment_years = np.random.exponential(scale=4, size=n)
active_loans = np.random.poisson(lam=1.5, size=n)
age = np.random.normal(loc=35, scale=8, size=n).clip(21, 65)

# Generate default probability using a logistic model
log_odds = (
    -2.0
    - 0.00003 * monthly_income
    + 0.000015 * outstanding_balance
    - 0.08 * employment_years
    + 0.35 * active_loans
    - 0.02 * age
)
prob_default = 1 / (1 + np.exp(-log_odds))
default = (np.random.random(n) < prob_default).astype(int)

credit = pd.DataFrame({
    'Monthly_Income': np.round(monthly_income, 0),
    'Outstanding_Balance': np.round(outstanding_balance, 0),
    'Employment_Years': np.round(employment_years, 1),
    'Active_Loans': active_loans,
    'Age': np.round(age, 0),
    'Default': default
})

print(f'Dataset 2: {credit.shape[0]} applicants × {credit.shape[1]} columns')
print(f'Default rate: {credit["Default"].mean():.1%}')
credit.head(10)
```
